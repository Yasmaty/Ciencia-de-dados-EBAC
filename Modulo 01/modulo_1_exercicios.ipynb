## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

# maiusculo: string.upper()
print(cancao.upper())

# posicao: string.find(substring)
posicao = cancao.find('moinho')
print(posicao)

# converter primeiro caracter para minusculo: casefold()
print(cancao.casefold())
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2.75% e supera expectativas; é a primeira alta em 6 anos.'

```

```


noticia = 'Selic vai a 2.75% e supera expectativas; é a primeira alta em 6 anos.'

# Extraindo o valor da taxa Selic
selic_inicio = noticia.find('Selic vai a')
selic_fim = noticia.find('%')
selic = noticia[selic_inicio + len('Selic vai a'):selic_fim + 1]

# Extraindo o valor do ano
ano_inicio = noticia.find('em')
ano_fim = noticia.find('anos')
ano = noticia[ano_inicio + len('em'):ano_fim].strip()

# Imprimindo os valores na tela
print("Taxa Selic:", selic)
print("Ano:", ano)
```
